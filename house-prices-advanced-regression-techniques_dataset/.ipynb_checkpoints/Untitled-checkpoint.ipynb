**Data Exploration and Cleaning**

1.We start by dropping dropping columns thaht are will have zero to no impact to our model

**HISTOGRAM**


```python
#plotting a histogram with matplotlib

plt.xlabel('House sale prices')
plt.hist(df['SalePrice'], bins=8)
plt.show()
```

The histogram s skewed to the right, thus we can conclude that majority of the house prices lie between 100000 to 2000000. 

we will need to carry out log transformation so that our model does not overfit the expensive houses or ignore them!

**Log Transformation**

```python
"""
Steps to carry out log transformation
1.Import Numpy
2.Create a new column that is required to store the column you will apply the log transformation on
3.Plot a histogram for the transformed colulmn(the new column)
"""

#log transfrmation on 'SalePrice'
df['log_sale_prices'] = np.log(df['SalePrice'])

#plotting a hist
plt.xlabel('Log transformed house prices')
plt.hist(df['log_sale_prices'], bins=8)
plt.show()
```

The histogram now is almost symetric, resembling a bell shape that a normal distribution should have.
The purpose of log transformation is to compress the large values that are in the expensive houses, much more than the small values in cheap houses, pulling the right tail inward and thus the bell shape

**FEATURE ENGINEERING AND ENCODING**

```python
df.info()
```

```python
"""
We want to know the age of the houses
We will create a variable house_age to store this information
"""

df['house_age'] = df['YrSold'] - df['YearBuilt']
#df['house_age']

df['year_since_remodelling'] = df['YrSold'] - df['YearRemodAdd']
#df['year_since_remodelling']

#dropping unnecessary columns
df = df.drop(columns=['YearRemodAdd', 'YearBuilt'])
df.info()
```

Dropping low signal columns basing it the heat map information
This include: salesprice - already replaced with log_sale_prices
poolArea, MiscVal, MoSold, YrSold

```python
df = df.drop(columns=['PoolArea', 'MiscVal', 'MoSold', 'YrSold', 'SalePrice'])
df.info()
```

**Encodoing Categorical Data**

```python
#Applying one hot encoding for categorical data

df_encoded = pd.get_dummies(df, drop_first=True)
df_encoded.info()
```

```python
#checking for colums whose data type is object
df_check = df_encoded.select_dtypes(include='object').columns
df_check
```

```python
#making the boolean columns integers
df_encoded = df_encoded = pd.get_dummies(df, drop_first=True, dtype=int)
df_encoded.info()
```

**THE TRAIN SPLITTING**


```python
#defining X and Y
X = df_encoded.drop(columns=['log_sale_prices'])
Y = df_encoded['log_sale_prices']
```

```python
#Importing the split function from sklearn
from sklearn.model_selection import train_test_split
```

```python
#importing the train_test_split function
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=50)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

**Plotting the Training and the Testing data in a scatter plot**

```python
#SCatter plots baby
#Scatter plot for training data
plt.scatter(X_train, Y_train, color='Pink', label='Traning scatter plot')
plt.legend()
plt.show()
```

```python
#Scatter plot for testing data
plt.scatter(X_test, Y_test, color='yellow', label='Testing scatter plot')
plt.legend()
plt.show()
```
